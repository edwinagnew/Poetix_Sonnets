```python
all_results = pd.read_csv('large_survey_1.csv')
all_results
```

```python
df = all_results[all_results['StartDate'] > '2022-01-12'].loc[2:]
df
```

```python
actual_codes = set(df['Random ID'])#.values
len(actual_codes), len(end_codes), len(actual_codes.intersection(end_codes))
```

```python
df2 = df[df['Random ID'].isin(end_codes)]
df2
```

```python
df_sane = df2.loc[(df2['Q1.2'] == 'G') & (df2['Q1.3'] == 'Their')].reset_index(drop=True)
```

```python
df_sane['random_sanity_correct'] = 1
for q, a in zip(['Q3.1', 'Q4.1', 'Q5.1', 'Q6.1'], ['No', 'Yes', 'No', 'Yes']):
    df_sane['random_sanity_correct'] *= (df_sane[q].isnull()) | (df_sane[q] == a)
df_sane = df_sane[df_sane['random_sanity_correct'] == 1]
df_sane['random_sanity_correct'].describe()
```

```python
lines_cols = [c for c in all_results.columns if 'How many lines' in all_results.loc[0, c]]
for lc in lines_cols:
    df_sane = df_sane[(df_sane[lc].isnull()) | (df_sane[lc] == '4')]
df_sane
```

```python
print(df_sane.columns)
```

```python
suffixes = {'theme': ['.2'],'poetic': ['.3_1'], 'grammar': ['.3_2'], 'emotion': ['.3_3'], 'lines': ['.3_4'], 'human': ['.3_5']}
```

```python
all_suffs = [c.split(".")[-1] for c in df_sane.columns if 'Q' in c]
set(all_suffs)
```

```python
human_cols = [c for c in all_results.columns if 'written by a human?' in all_results.loc[0, c]]
print(human_cols)
len(human_cols)
```

```python
def flatten(xs):
    return [item for sublist in xs for item in sublist]
```

```python
authors = ['hafez', 'deepspeare', 'peter', 'bill', 'sylvia', 'human']
all_scores = {}
```

```python
human_scores = {}
auth_cols = {}

for hc in human_cols:
    q_num = int(hc.strip('Q').split('.')[0])
    num = (q_num - 7)//25
    
    auth = authors[num]
    if auth not in human_scores: 
        human_scores[auth] = []
        auth_cols[auth] = []
    
    auth_cols[auth].append(q_num)
        
    scores = df_sane[hc].dropna().astype('int').values
        
    human_scores[auth].append(list(scores))
print({k:sum([len(w) for w in v]) for k,v in human_scores.items()})
all_scores['human'] = human_scores
```

```python
fig, axs = plt.subplots(1, len(authors), figsize=(5*len(authors), 5))

for i, aut in enumerate(authors):
    scores = flatten(human_scores[aut])

    axs[i].hist(scores)
    m = np.mean(scores)
    sd = np.std(scores)
    axs[i].set_title("humanness: " + aut + "\nmean=" + str(np.round(m, 3)) + "\nsd=" + str(np.round(sd, 3)))
```

```python
poetic_scores = {}
poetic_cols = [c for c in all_results.columns if 'poetic does this sound?' in all_results.loc[0, c]]
print(len(poetic_cols))
for c in poetic_cols:
    q_num = int(c.strip('Q').split('.')[0])
    num = (q_num - 7)//25
    
    auth = authors[num]
    if auth not in poetic_scores: 
        poetic_scores[auth] = []
        
    scores = df_sane[c].dropna().astype('int').values
        
    poetic_scores[auth].append(list(scores))
print({k:len(v) for k,v in poetic_scores.items()})
all_scores['poetic'] = poetic_scores
```

```python
fig, axs = plt.subplots(1, len(authors), figsize=(5*len(authors), 5))

for i, aut in enumerate(authors):
    scores = flatten(poetic_scores[aut])

    axs[i].hist(scores)
    m = np.mean(scores)
    sd = np.std(scores)
    axs[i].set_title("poeticness: " + aut + "\nmean=" + str(np.round(m, 3)) + "\nsd=" + str(np.round(sd, 3)))
```

```python
grammar_scores = {}
grammar_cols = [c for c in all_results.columns if 'rate how grammatical' in all_results.loc[0, c]]
print(len(grammar_cols))
for c in grammar_cols:
    q_num = int(c.strip('Q').split('.')[0])
    num = (q_num - 7)//25
    
    auth = authors[num]
    if auth not in grammar_scores: 
        grammar_scores[auth] = []
        
    scores = df_sane[c].dropna().astype('int').values
        
    grammar_scores[auth].append(list(scores))
print({k:len(v) for k,v in grammar_scores.items()})
all_scores['grammar'] = grammar_scores
```

```python
fig, axs = plt.subplots(1, len(authors), figsize=(5*len(authors), 5))

for i, aut in enumerate(authors):
    scores = flatten(grammar_scores[aut])

    axs[i].hist(scores)
    m = np.mean(scores)
    sd = np.std(scores)
    axs[i].set_title("grammar: " + aut + "\nmean=" + str(np.round(m, 3)) + "\nsd=" + str(np.round(sd, 3)))
```

```python
emotion_scores = {}
emotion_cols = [c for c in all_results.columns if 'conveys emotion' in all_results.loc[0, c]]
print(len(emotion_cols))
for c in emotion_cols:
    q_num = int(c.strip('Q').split('.')[0])
    num = (q_num - 7)//25
    
    auth = authors[num]
    if auth not in emotion_scores: 
        emotion_scores[auth] = []
        
    scores = df_sane[c].dropna().astype('int').values
        
    emotion_scores[auth].append(list(scores))
print({k:len(v) for k,v in emotion_scores.items()})
all_scores['emotion'] = emotion_scores
```

```python
fig, axs = plt.subplots(1, len(authors), figsize=(5*len(authors), 5))

for i, aut in enumerate(authors):
    scores = flatten(emotion_scores[aut])

    axs[i].hist(scores)
    m = np.mean(scores)
    sd = np.std(scores)
    axs[i].set_title("emotion: " + aut + "\nmean=" + str(np.round(m, 3)) + "\nsd=" + str(np.round(sd, 3)))
```

```python
lines_scores = {}
lines_cols = [c for c in all_results.columns if 'How many lines' in all_results.loc[0, c]]
print(len(lines_cols))
for c in lines_cols:
    q_num = int(c.strip('Q').split('.')[0])
    num = (q_num - 7)//25
    
    auth = authors[num]
    if auth not in lines_scores: 
        lines_scores[auth] = []
        
    scores = df_sane[c].dropna().astype('int').values
        
    lines_scores[auth].append(list(scores))
print({k:len(v) for k,v in lines_scores.items()})
```

```python
fig, axs = plt.subplots(1, len(authors), figsize=(5*len(authors), 5))

for i, aut in enumerate(authors):
    scores = flatten(lines_scores[aut])

    
    axs[i].hist(scores)
    m = np.mean(scores)
    sd = np.std(scores)
    axs[i].set_title("lines: " + aut + "\nmean=" + str(np.round(m, 3)) + "\nsd=" + str(np.round(sd, 3)))
```

```python
xs = np.arange(len(authors))

types = ['human', 'emotion', 'poetic']
fig, axs = plt.subplots(1, len(types), figsize=(5*len(types), 5))

for i, t in enumerate(types):
    aut_scores = [np.mean(flatten(all_scores[t][aut])) for aut in authors]
    #axs[i].bar(xs, [np.mean(all_scores[t][aut]) for aut in authors], width=0.2)
    axs[i].bar(xs, aut_scores, width=0.2)
    
    axs[i].set_xticks(xs + 0.2 // 2)
    axs[i].set_xticklabels(authors)
    axs[i].set_title(t)
```
