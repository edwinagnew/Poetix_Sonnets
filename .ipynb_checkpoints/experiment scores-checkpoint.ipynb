```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/Users/edwinagnew/Downloads/Experiment Sonnet Stanzas - Sheet1.csv')
df = df.drop([0, 1, 2, 3, 4])
df_scores = df.copy()
df
```

```python
import scenery
sc = scenery.Scenery_Gen()
```

```python
import gpt_2
sc.gpt =  gpt_2.gpt_gen(sonnet_object=sc)
```

```python
df_scores = df_scores.reset_index()
df_scores
```

```python
poets = ["Bill Turing", "Sylvia", "Peter's Code (Poetix 2016)", "Hafez", "Deepspeare", "Human (no theme)"]
for poet in poets:
    for i, poem in enumerate(df[poet]):
        if type(poem) == float: df_scores.loc[i, poet] = np.nan
        else:
            poem = poem.replace("\n\n", "\n")
            print(i)
            print(poem)
            df_scores.loc[i, poet] = np.mean(sc.gpt.score_line(poem.strip().split("\n")))
            print(df_scores.loc[i, poet], "\n")
```
