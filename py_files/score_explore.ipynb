```python
import matplotlib.pyplot as plt
%matplotlib notebook

x = scores[:,1] #x is second line
y = scores[:,2] #y is whole thing
names = np.array([s[0] for s in couplets])
c = scores[:,0] #c is first line

#norm = plt.Normalize(1,4)
cmap = plt.cm.RdYlGn

fig,ax = plt.subplots()
sc = plt.scatter(x,y,c=c, s=100, cmap=cmap)#, norm=norm)
plt.xlabel('second line')
plt.ylabel('whole couplet')
annot = ax.annotate("", xy=(0,0), xytext=(20,20),textcoords="offset points",
                    bbox=dict(boxstyle="round", fc="w"),
                    arrowprops=dict(arrowstyle="->"))
annot.set_visible(False)

def update_annot(ind):

    pos = sc.get_offsets()[ind["ind"][0]]
    print(pos)
    print(sc.get_offsets())
    annot.xy = pos
    text = "{}".format(" ".join([names[n] for n in ind["ind"]]))
    annot.set_text(text)
    annot.get_bbox_patch().set_facecolor(cmap(c[ind["ind"][0]]))
    annot.get_bbox_patch().set_alpha(1)


def hover(event):
    vis = annot.get_visible()
    if event.inaxes == ax:
        cont, ind = sc.contains(event)
        if cont:
            update_annot(ind)
            annot.set_visible(True)
            fig.canvas.draw_idle()
        else:
            if vis:
                annot.set_visible(False)
                fig.canvas.draw_idle()

fig.canvas.mpl_connect("motion_notify_event", hover)
plt.xlabel('second template loss')
plt.ylabel('whole line loss')
#print(sc.get_offsets(), "d")
plt.show()
```
